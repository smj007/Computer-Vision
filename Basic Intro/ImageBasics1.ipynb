```python
img_bgr = cv2.imread('../DATA/00-puppy.jpg')
plt.imshow(img_bgr)
```

```python
img_rgb = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2RGB)
plt.imshow(img_rgb)
```

```python
img_gray = cv2.imread('../DATA/00-puppy.jpg', cv2.IMREAD_GRAYSCALE)
plt.imshow(img_gray)

#grayscale color is diff here due to matplotlib color scheme - VIRIDIS
#here the darker text (0) is dark blue and lighter (~255) is greenish
```

```python
#use cmap to gray
img_gray = cv2.imread('../DATA/00-puppy.jpg', cv2.IMREAD_GRAYSCALE)
plt.imshow(img_gray, cmap = 'gray')
```

```python
#resize 
#main point - we specify WHC but the size display is always in HWC
img_rgb.shape
```

```python
img_resized = cv2.resize(img_rgb, (1300, 300))  #both have been resized
```

```python
plt.imshow(img_resized)
```

```python
w_ratio = 0.8
h_ratio = 0.2
```

```python
img_rgb.shape
```

```python
new_img = cv2.resize(img_rgb, (0, 0), img_rgb, w_ratio, h_ratio)
```

```python
plt.imshow(new_img)
```

```python
new_img.shape
```

```python
#flipping images - axis == 0 Hor 1 Ver -1 Both
new_img = cv2.flip(img_gray, 0)
plt.imshow(new_img)
```

```python
new_img = cv2.flip(img_gray, 1)
plt.imshow(new_img)
```

```python
new_img = cv2.flip(img_gray, -1)
plt.imshow(new_img)
```

```python
#saving an image to dir
type(new_img)
```

```python
cv2.imwrite('new_picture.jpg', new_img)
```
