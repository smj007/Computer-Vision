```python
img = cv2.imread('../DATA/rainbow.jpg', 0)
plt.imshow(img, cmap = 'gray')
```

```python
ret, th1 = cv2.threshold(img, 127, 255, cv2.THRESH_BINARY)
plt.imshow(th1, cmap = 'gray') #ret returns the th value - 127
```

```python
ret, th2 = cv2.threshold(img, 127, 255, cv2.THRESH_BINARY_INV)
plt.imshow(th2, cmap = 'gray')  # < 127 to 1 and >127 to 0 
```

```python
ret, th3 = cv2.threshold(img, 127, 255, cv2.THRESH_TRUNC)
plt.imshow(th3, cmap = 'gray') #trunc - all above 127 are 127 , rest all as it is
```

```python
ret, th4 = cv2.threshold(img, 200, 255, cv2.THRESH_TOZERO)
plt.imshow(th4, cmap = 'gray')  #all those above threshold remain as it is, rest of it below are pushed to 0
```

```python
#apply to a vintage pic

img = cv2.imread("../DATA/crossword.jpg",0)
```

```python
def show_pic(img):
    fig = plt.figure(figsize = (15, 15))
    ax = fig.add_subplot(111)
    ax.imshow(img, cmap = 'gray')
```

```python
show_pic(img)
```

```python
#binary
res, th1 = cv2.threshold(img, 127, 255, cv2.THRESH_BINARY)
show_pic(th1)
```

```python
res, th2 = cv2.threshold(img, 160, 255, cv2.THRESH_BINARY)
show_pic(th2)
```

```python
#ADAPTIVE THRESHOLDING
th3 = cv2.adaptiveThreshold(img, 255, cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY, 11, 8)
#     .   @param blockSize Size of a pixel neighborhood that is used to calculate a threshold value for the
#     .   pixel: 3, 5, 7, and so on.
#     .   @param C Constant subtracted from the mean or weighted mean (see the details below). Normally, it
#     .   is positive but may be zero or negative as well.

#only one return val here
```

```python
show_pic(th3)
```

```python
th4 = cv2.adaptiveThreshold(img, 255, cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY, 15, 8)
show_pic(th4)
```

```python
blended = cv2.addWeighted(src1 = th1, alpha = 0.7, src2 = th4, beta = 0.3, gamma = 0)
show_pic(blended)
```
