```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
img = cv2.imread('../DATA/dog_backpack.png')
```

```python
plt.imshow(img)
```

```python
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img_rgb)
```

#### TASK: Flip the image upside down and display it in the notebook.

```python
img_copy = img_rgb.copy()
img_copy = cv2.flip(img_copy, 0)
plt.imshow(img_copy)
```

#### TASK: Draw an empty RED rectangle around the dogs face and display the image in the notebook.

```python
pt1 = (225, 375)
pt2 = (600, 725)
cv2.rectangle(img_rgb, pt1 = pt1, pt2 = pt2, color = (255, 0, 0), thickness = 10)
plt.imshow(img_rgb)
```

```python
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
vertices = np.array([[225, 725],[475, 350],[700, 725]], np.int32)
pts = vertices.reshape((-1, 1, 2))

cv2.polylines(img_rgb, [pts], isClosed = True, color = (0, 0, 255), thickness = 15)
plt.imshow(img_rgb)
```

```python
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
vertices = np.array([[225, 725],[475, 350],[700, 725]], np.int32)
pts = vertices.reshape((-1, 1, 2))

cv2.fillPoly(img_rgb,[pts],color=(0,0,255))
plt.imshow(img_rgb)
```

```python
import cv2
import numpy as np

def draw_circle(event, x, y, flags, param):
    
    if event == cv2.EVENT_RBUTTONDOWN:
        cv2.circle(img, (x, y), 100, (0, 0, 255), 10)

img = cv2.imread("../DATA/dog_backpack.png")
cv2.namedWindow(winname = 'doggie')
cv2.setMouseCallback('doggie', draw_circle)

while True:
    cv2.imshow('doggie', img)
    if cv2.waitKey(1) & 0xFF == 27:
        break
        
cv2.destroyAllWindows()           
```
